```python
""" Santigrat ve Fahrenhayt cinsinden sıcaklıkları dönüştürmek
için bir Python programı yazın.
• c/5 = f-32/9 (c , santigrat cinsinden sıcaklık; f, fahrenhayt
cinsinden sıcaklık)
"""
sicaklik = input("sicaklık girin:")
derece =int(sicaklik[:-1])
tip=sicaklik[-1]
match tip:
    case "c":
        result=(9 * derece)/5 +32
        print(derece,"c= ",result,"f dir")
    case "f":
        result=(derece - 32)*5 /9
        print(derece,"c= ",result,"f dir")
    case _:
        print("uyumsuz giriş")
```

```python
"""return (isteğe bağlı) - bir fonksiyondan değer
döndürür"""
def sayilariTopla(sayi1,sayi2):
    toplam=sayi1 +sayi2
    print("toplam:",toplam)
sayilariTopla(5,7)
```

```python
def sayilariTopla(sayi1,sayi2):
    toplam=sayi1 +sayi2
    print("toplam:",toplam)
sayilariTopla(sayi2=5,sayi1=7)
```

```python
def sayilariTopla(sayi1=5,sayi2=7):
    toplam=sayi1 +sayi2
    print("toplam:",toplam)
sayilariTopla(3,2)
sayilariTopla(sayi2=15)
sayilariTopla()
```

```python
def sayilariTopla(*sayilar):
    sonuc=0
    for sayi in sayilar:
        sonuc = sonuc + sayi
    print("toplam:",sonuc)
sayilariTopla(1,2,3)
sayilariTopla(4,9)
sayilariTopla(7)
sayilariTopla()
```

```python
"""Eğer fonksiyonumuzun bir fonksiyon çağrısına bir
değer döndürmesini istiyorsak, "return" deyimini
kullanırız.

"return" deyimi aynı zamanda işlevin sona
erdiğini de belirtir. Dönüşten sonra herhangi bir
kod yürütülmez."""

def sayilariTopla(sayi1,sayi2):
    toplam=sayi1 +sayi2
    return toplam
print(sayilariTopla(3,5))
```

```python
def KareHesapla(sayi):
    sonuc=sayi**2
    return sonuc
kare=KareHesapla(9)
print(kare)
```
